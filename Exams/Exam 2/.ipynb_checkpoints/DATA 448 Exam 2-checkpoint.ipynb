```python
#Replacing na values in profession ((SAME AS EXAM 1))
#Identifying categorical variables
categorical_vars = customers.select_dtypes(include = 'object').columns.to_list()
print(categorical_vars)

frequent_values = customers[categorical_vars].mode().iloc[0].to_dict()
print(frequent_values)

## Filling missing values with most frequent labels
customers = customers.fillna(value = frequent_values)

#Identifying numeric variables
numeric_vars = customers.select_dtypes(exclude = 'O').columns.to_list()
print(numeric_vars)

#Choosing median because the data isn't normal
#Compute the median values of the numerical variables and store in a dictionary
median_values = customers[numeric_vars].median().to_dict()
median_values

#Let's replace missing data with the median
customers = customers.fillna(value = median_values)
customers.isna().sum()
```

(c) (5 points) Split the customers data-frame into: train (80%) and test (20%).


```python
'''
Descending order:
Annual Income/Family Size    1.00
interaction_1                0.99
Annual Income_squared        0.89
Age                          0.68
interaction_2                0.45
Profession                   0.22
Work Experience              0.17
TOP 7^^^
Family Size                  0.13
Gender                       0.03
'''
### Bringing in the engineered variables from exam 1 
##Engineered varaible 1
customers['interaction_1'] = customers['Annual Income ($)'] * customers['Age']

##Engineered variable 2
customers['interaction_2'] = customers['Annual Income ($)'] * customers['Work Experience']

##Engineered variable 3
customers['Annual Income/Family Size'] = customers['Annual Income ($)']/ customers['Family Size']

##Engineered variable 4
customers['Annual Income_squared'] = customers['Annual Income ($)']**2

##Engineered variable 5
customers['Gender'] = np.where(customers['Gender'] == 'Male', 0, 1)

##Engineered variable 6
customers['Profession'] = np.where(customers['Profession'] == 'Healthcare',1,
                                 np.where(customers['Profession'] == 'Engineer',2,
                                 np.where(customers['Profession'] == 'Lawyer',3,
                                 np.where(customers['Profession'] == 'Entertainment',4,
                                 np.where(customers['Profession'] == 'Artist',5,
                                 np.where(customers['Profession'] == 'Executive',6,
                                 np.where(customers['Profession'] == 'Doctor',7,
                                 np.where(customers['Profession'] == 'Homemaker',8,9))))))))
customers.head()
```

(d) (20 points) Using results from Exercise 16 part (e) in Exam 1, use the top 7 variables,
build a model on the train data-frame, and tune this model using the Optuna framework. Notice that you need to tune the model based on the root mean squared error.
Then, use the optimal model to make predictions on the test data-frame. Finally,
compute the root mean squared error of the model. Make sure to run at least 50 trials
the Optuna framework

```python
### RANDOM FOREST ###

class Objective:
    def __init__(self, seed):
        self.seed = seed
        
    def __call__(self, trial):
        params = dict(n_estimators = trial.suggest_int('n_estimators', 100, 2000),
                        min_samples_split = trial.suggest_int('min_samples_split', 5, 30),
                        min_samples_leaf = trial.suggest_int('min_samples_leaf', 5, 30),
                        max_depth = trial.suggest_int('max_depth', 2, 10)
                        )
    
        scores = list()
        
        skf = KFold(n_splits = 3, shuffle = True, random_state = self.seed)
        
        for train_idx, valid_idx in skf.split(X_train, Y_train):
            X_train1, X_valid1 = X_train.iloc[train_idx], X_train.iloc[valid_idx]
            Y_train1, Y_valid1 = Y_train.iloc[train_idx], Y_train.iloc[valid_idx]
            
            RF_md = RandomForestRegressor(**params).fit(X_train1, Y_train1)
            pred_valid = RF_md.predict(X_valid1)   
            score = mean_squared_error(Y_valid1, pred_valid)
            scores.append(score)
            
        return np.mean(scores)
```

```python
SEED = 42
N_TRIALS = 50

##Executing optuna
study = optuna.create_study(direction = 'minimize')
study.optimize(Objective(SEED), n_trials = N_TRIALS)
```

```python
print('The best parameter combo with RandomForestRegressor is:', study.best_trial.params) 
```

```python
## RandomForest Regressor with best hyperparameters
RF_md = RandomForestRegressor(**study.best_trial.params).fit(X_train, Y_train)

##predicting on test
RF_pred = RF_md.predict(X_test)

##computing mse
RF_results = mean_squared_error(Y_test, RF_pred, squared = False)
print('The RMSE of the RF model with the best hyperparameters:',RF_results)
```

```python
### XGBOOST ###

class Objective: 
    def __init__(self, seed):    
        self.seed = seed
        
    def __call__(self, trial):  
        params = dict(n_estimators = trial.suggest_int('n_estimators', 100, 2000),
                      max_depth = trial.suggest_int('max_depth', 2, 10),
                      min_child_weight = trial.suggest_int('min_child_weight', 2, 20),
                      learning_rate = trial.suggest_float('learning_rate', 0.001, 0.1, log = True),
                      gamma = trial.suggest_float('gamma', 0.0, 10.0),
                      colsample_bytree = trial.suggest_float('colsample_bytree', 0.2, 0.9),
                      subsample = trial.suggest_float('subsample', 0.2, 0.9))
        
        scores2 = list()
        
        skf = KFold(n_splits = 3, shuffle = True, random_state = self.seed)
        
        for train_idx, valid_idx in skf.split(X_train, Y_train):
            
            X_train2, X_valid2 = X_train.iloc[train_idx], X_train.iloc[valid_idx]
            Y_train2, Y_valid2 = Y_train.iloc[train_idx], Y_train.iloc[valid_idx]
            
            XGB_md = XGBRegressor(**params).fit(X_train2, Y_train2) 
            pred_valid2 = XGB_md.predict(X_valid2)
            
            score2 = mean_squared_error(Y_valid2, pred_valid2, squared = False)
            scores2.append(score2)
            
        return np.mean(scores2)
```

```python
### GRADIENT BOOSTING ### 

class Objective:
    def __init__(self, seed):
        self.seed = seed
        
    def __call__(self, trial):
        params = dict(n_estimators = trial.suggest_int('n_estimators', 100, 2000),
                    min_samples_split = trial.suggest_int('min_samples_split', 5, 30),
                    min_samples_leaf = trial.suggest_int('min_samples_leaf', 5, 30),
                    max_depth = trial.suggest_int('max_depth', 2, 10),
                    learning_rate= trial.suggest_uniform('learning_rate', .001,.01)
                    )
    
        scores3 = list()
        
        skf3 = KFold(n_splits = 3, shuffle = True, random_state = self.seed)
        
        for train_idx, valid_idx in skf3.split(X_train, Y_train):
            X_train3, X_valid3 = X_train.iloc[train_idx], X_train.iloc[valid_idx]
            Y_train3, Y_valid3 = Y_train.iloc[train_idx], Y_train.iloc[valid_idx]
            
            GB_md = GradientBoostingRegressor(**params).fit(X_train3, Y_train3)
            pred_valid3 = GB_md.predict(X_valid3)   
            score3 = mean_squared_error(Y_valid3, pred_valid3)
            scores3.append(score3)
            
        return np.mean(scores3)
```
